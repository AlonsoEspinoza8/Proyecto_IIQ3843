# Full plots for the $NH_3$ simulation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_data(file_path):
    """Load simulation data from a text file."""
    data = np.loadtxt(file_path, delimiter=',', skiprows=1)
    return data

flow_rate_0 = load_data('ammonia_tank_full_data_mL_0.0e-6_kg_per_s.txt')
flow_rate_1 = load_data('ammonia_tank_full_data_mL_1.0e-6_kg_per_s.txt')
flow_rate_5 = load_data('ammonia_tank_full_data_mL_5.0e-6_kg_per_s.txt')
flow_rate_10 = load_data('ammonia_tank_full_data_mL_10.0e-6_kg_per_s.txt')
flow_rate_25 = load_data('ammonia_tank_full_data_mL_25.0e-6_kg_per_s.txt')
```

First plot: Plot the $\dot{B}_L$ and $\dot{B}$ for each flow rate

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# --- Primer subplot: B_l [7] ---
axes[0].plot(flow_rate_0[0], flow_rate_0[7], label=rf'$m_L=0.0\cdot 10^{{-{6}}}$ $kg/s$')
axes[0].plot(flow_rate_1[0], flow_rate_1[7], label=rf'$m_L=1.0\cdot 10^{{-{6}}}$ $kg/s$')
axes[0].plot(flow_rate_5[0], flow_rate_5[7], label=rf'$m_L=5.0\cdot 10^{{-{6}}}$ $kg/s$')
axes[0].plot(flow_rate_10[0], flow_rate_10[7], label=rf'$m_L=10.0\cdot 10^{{-{6}}}$ $kg/s$')
axes[0].plot(flow_rate_25[0], flow_rate_25[7], label=rf'$m_L=25.0\cdot 10^{{-{6}}}$ $kg/s$')
axes[0].set_title(r'$ \dot{B}_L $ vs Time')
axes[0].set_xlabel('Time (s)')
axes[0].set_ylabel('Boil-off Rate (kg/s)')
axes[0].legend()
axes[0].grid()

# --- Segundo subplot: B [8] ---
axes[1].plot(flow_rate_0[0], flow_rate_0[8], label=rf'$m_L=0.0\cdot 10^{{-{6}}}$ $kg/s$')
axes[1].plot(flow_rate_1[0], flow_rate_1[8], label=rf'$m_L=1.0\cdot 10^{{-{6}}}$ $kg/s$')
axes[1].plot(flow_rate_5[0], flow_rate_5[8], label=rf'$m_L=5.0\cdot 10^{{-{6}}}$ $kg/s$')
axes[1].plot(flow_rate_10[0], flow_rate_10[8], label=rf'$m_L=10.0\cdot 10^{{-{6}}}$ $kg/s$')
axes[1].plot(flow_rate_25[0], flow_rate_25[8], label=rf'$m_L=25.0\cdot 10^{{-{6}}}$ $kg/s$')
axes[1].set_title(r'$\dot{B}$ vs Time')
axes[1].set_xlabel('Time (s)')
axes[1].legend()
axes[1].grid()

# Ajustar diseo
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(flow_rate_0[0], flow_rate_0[8], label=rf'$m_L=0.0\cdot 10^{{-{6}}}$ $kg/s$')
plt.plot(flow_rate_1[0], flow_rate_1[8], label=rf'$m_L=1.0\cdot 10^{{-{6}}}$ $kg/s$')
plt.plot(flow_rate_5[0], flow_rate_5[8], label=rf'$m_L=5.0\cdot 10^{{-{6}}}$ $kg/s$')
plt.plot(flow_rate_10[0], flow_rate_10[8], label=rf'$m_L=10.0\cdot 10^{{-{6}}}$ $kg/s$')
plt.plot(flow_rate_25[0], flow_rate_25[8], label=rf'$m_L=25.0\cdot 10^{{-{6}}}$ $kg/s$')

plt.ylabel('Boil-off Rate (kg/s)')
plt.title('Boil-off Rates vs Time for Different Mass Flow Rates')
plt.legend()
plt.grid()
plt.show()  
```

Second plot: Plot the mass of the vapour inside the tank

```python
# Plot the vapour height and the liquid height for the 25e-6 kg/s flow rate
d_i = 0.201
internal_area = np.pi * (d_i / 2) ** 2
tank_volume = 6.75e-3  # m^3
tank_height = tank_volume / internal_area

liquid_height_25 = np.array(flow_rate_25[6]) / internal_area
vapour_height_25 = tank_height - liquid_height_25


plt.figure(figsize=(12, 6))
plt.plot(np.array(flow_rate_25[0])/3600, liquid_height_25, label='Liquid Height')
plt.plot(np.array(flow_rate_25[0])/3600, vapour_height_25, label='Vapour Height')
plt.axhline(y=tank_height, color='black', linestyle='--', label='Tank Height')
plt.ylabel(r'Height ($m$)')
plt.xlabel(r'Time (hours)')
plt.title(r'Liquid and Vapour Heights vs Time for $m_L=25.0\cdot 10^{-6}$ $kg/s$')
plt.legend()
plt.grid()
plt.show()
```

Third plot: Graficate the BOG temperature for all flow rates

```python
# Calculate the BOG temperature for all flow rates
plt.figure(figsize=(12, 6))
plt.plot(flow_rate_0[0], flow_rate_0[11], label=rf'$m_L=0.0\cdot 10^{{-{6}}}$ $kg/s$')
plt.plot(flow_rate_1[0], flow_rate_1[11], label=rf'$m_L=1.0\cdot 10^{{-{6}}}$ $kg/s$')
plt.plot(flow_rate_5[0], flow_rate_5[11], label=rf'$m_L=5.0\cdot 10^{{-{6}}}$ $kg/s$')
plt.plot(flow_rate_10[0], flow_rate_10[11], label=rf'$m_L=10.0\cdot 10^{{-{6}}}$ $kg/s$')
plt.plot(flow_rate_25[0], flow_rate_25[11], label=rf'$m_L=25.0\cdot 10^{{-{6}}}$ $kg/s$')
plt.ylabel(r'Boil-off Temperature ($K$)')
plt.title('Boil-off Temperature vs Time for Different Mass Flow Rates')
plt.legend()
plt.grid()
plt.show()  

# Calculate the BOG temperature for all flow rates
plt.figure(figsize=(12, 6))
plt.plot(flow_rate_0[0], flow_rate_0[1], label=rf'$m_L=0.0\cdot 10^{{-{6}}}$ $kg/s$')
plt.plot(flow_rate_1[0], flow_rate_1[1], label=rf'$m_L=1.0\cdot 10^{{-{6}}}$ $kg/s$')
plt.plot(flow_rate_5[0], flow_rate_5[1], label=rf'$m_L=5.0\cdot 10^{{-{6}}}$ $kg/s$')
plt.plot(flow_rate_10[0], flow_rate_10[1], label=rf'$m_L=10.0\cdot 10^{{-{6}}}$ $kg/s$')
plt.plot(flow_rate_25[0], flow_rate_25[1], label=rf'$m_L=25.0\cdot 10^{{-{6}}}$ $kg/s$')
plt.ylabel(r'Average vapour temperature ($K$)')
plt.title('Average Vapour Temperature vs Time for Different Mass Flow Rates')
plt.legend()
plt.grid(True)
plt.show()  
```
